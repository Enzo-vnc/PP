```python
### RDKit part 2 

### Structure matching 
"""
Find specific structure in a molecule. 
function from RDKit 'HasSubstructMatch', which returns True if the molecule contains the substructure and False 
otherwise. 
Or 'GetSubstructMatch', which returns the indices of the atoms in the molecule that match the substructure.
"""
# Let's start by loading the RDKit library and the drug molecules from last week's exercise : 

# some imports
from rdkit import Chem
from rdkit.Chem import PandasTools
import pandas as pd
from pathlib import Path
import os

current_file = Path(os.path.abspath(''))
csv_file = current_file.parent / "PP" / "chembl_drugs.csv"
df = pd.read_csv(csv_file, sep= ";")
df.head()
```

```python
# remove the rows with NaN values
df = df.dropna(subset=["Smiles"])
```

```python
# add molecule objects to the df from SMILES in a column named "mol" using RDKit functions 
#(PandasTools or Chem module)
PandasTools.AddMoleculeColumnToFrame(df, smilesCol="Smiles", molCol="mol")
```

```python
# if PandasTools is not wokring in you RDKit version:
df["mol"] = df["Smiles"].apply(Chem.MolFromSmiles)
```

```python
## Tyrosina Kinase Inhibitors

# First, let's get the motif that binds the hinge region of tyrosine kinases: 
indazole_smi = "c1cc2c(cc1)C=NN2"
indazole = Chem.MolFromSmiles(indazole_smi)
indazole
```

```python
# Let's test the HasSubstructMatch function on one example molecule to see how it works.
test_mol = df["mol"][0]
test_mol
```

```python
test_mol.HasSubstructMatch(indazole) 
```

```python
# Now, let's query all the drug molecules for the presence of the hinge binding motif. 
#use the HasSubstructMatch function in combinations with pandas.apply to check if the molecule contains 
#the motif.
df['has_indazole'] = df['mol'].apply(lambda x: x.HasSubstructMatch(indazole))
```

```python
# SAN Definitions roughly tell us what class of drug we are looking at. We'll remove missing values 
#for analysis purposes.

# remove rows that do not have a USAN definition
df = df.dropna(subset=["USAN Definition"])
df.head()
```

```python
#Filter the molecules:
#1)remove the molecules that do not contain the motif
#2)only keep molecules that are tyrosine kinase inhibitors

# 1) only keep those with the motif
df_fil = df[df['has_indazole']]
df_fil.head()
```

```python
# extract all USAN Definitions
moa = df_fil['USAN Definition'].unique().tolist()
moa
```

```python
# counts of moa in df_fil
df_fil['USAN Definition'].value_counts()
```

```python
# In this list by counts of molecules, we can see that anticancer drugs (antineoplastics) are the most 
#common but also "tyrosine kinase inhibitors" are frequent. They appear with different names. Filter the 
#dataframe so that the row USAN Definitions only needs to contain the string but does not need to match 
#completely.

# 2) only keep tyrosine kinase inhibitors
tki = df_fil[df_fil['USAN Definition'].str.contains("tyrosine kinase inhibitor")]
tki
```

```python
# visualize the remaining TKI structures using the PandasTools.FrameToGridImage function
PandasTools.FrameToGridImage(tki, column='mol', legendsCol="USAN Definition", molsPerRow=5, subImgSize=(300, 300))
```

```python
# The function GetSubstructMatch returns the indices of the atoms in the molecule that match the substructure. 
#Let's use this function to visualize the substructure in the first molecule.

mol = tki['mol'].iloc[0]
match_indices = mol.GetSubstructMatch(indazole)
match_indices
```

```python
#  highlight the indazole motif using the Draw.MolToImage function
from rdkit.Chem import Draw
img = Draw.MolToImage(mol, highlightAtoms=match_indices)
img
```

```python
## Thalidomide case 

s_thalidomide = Chem.MolFromSmiles('O=C1c2ccccc2C(=O)N1[C@H]3CCC(=O)NC3=O')
s_thalidomide
```

```python
# find the Thalidomide molecule object in the dataframe
thalidomide = df[df["Name"] == "THALIDOMIDE"]['mol'].values[0]
type(thalidomide)
```

```python
# check if the Thalidomide molecule object has the S motif using the HasSubstructMatch function
thalidomide.HasSubstructMatch(s_thalidomide)
```

```python
thalidomide
```

```python
"""
It tells us that the molecule does contain the s-enantiomer. However, when we look at the molecule, we see 
that actually no stereochemistry is defined. RDKit does not consider stereochemistry by default when searching 
for substructures. We can change this by setting the useChirality parameter to True. Let's see if we can find 
the s-enantiomer of thalidomide in our database.
"""
# check if thalidomide contains the S-enantiomer by enforcing chirality
thalidomide.HasSubstructMatch(s_thalidomide, useChirality=True)
```
